```python
# 계좌 클래스 
class Account:
    # 기본 생성자 (생략 가능함)
    def __init__(self):
        pass
    
    '''
      self.__balance
            ㄴ double underline (dounder)
            ㄴ private member variable
                ㄴ 같은 클래스 안에서만 접근 가능하고
                   외부에서는 접근할 수 없음
    '''
    
    # 계좌 개설하기
    def account_open(self, money):
        print('축하합니다. 새 계좌를 개설하셨습니다')
        self.__balance = money
        self.display_account()
        
    # 입금하기 (setter)    
    def deposit(self, money):
        if money > 0:
            self.__balance += money
        else:
            print('금액을 다시 확인해 주세요')
        self.display_account()
        
    # 출금하기 (setter)    
    def withdraw(self, money):
        if money < self.__balance:
            self.__balance -= money
        else:
            print('잔액이 부족합니다')
        self.display_account()
        
    # 계좌 잔액 확인하기  (getter)
    def display_account(self):
        print(f"현재 잔액 : {format(self.__balance, ',')} 원")
```

```python
acc1 = Account()
acc1.account_open(10000)

acc1.deposit(20000)
acc1.deposit(50000)
acc1.deposit(10000)

acc1.withdraw(30000)

acc1.deposit(-20000)

acc1.withdraw(60000)

acc1.balance = 100000000
acc1.display_account()
```

```python
# 4 가지 계산 기능이 있는 클래스
class Calc4:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        self.result = 0
    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def multiply(self):
        return self.num1 * self.num2
    def divide(self):
        return self.num1 / self.num2
```

```python
c1 = Calc4(10, 8)
print('c1.num1 :',c1.num1)
print('c1.num2 :',c1.num2)
print(c1.add())
print(c1.subtract())
print(c1.multiply())
print(c1.divide())
```

```python
'''
4 가지 계산은 기본으로 하고
제곱연산 기능이 추가된 
클래스 작성하기
'''
class Calc4_1:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        self.result = 0
    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def multiply(self):
        return self.num1 * self.num2
    def divide(self):
        return self.num1 / self.num2
    def power(self):
        return self.num1 ** self.num2
```

```python
c1_1 = Calc4_1(10, 8)
print(c1_1.num1)
print(c1_1.num2)

print(c1_1.add())
print(c1_1.subtract())
print(c1_1.multiply())
print(c1_1.divide())
print(c1_1.power())
```

```python
'''
4 가지 계산은 기본으로 하고
제곱연산 기능이 추가된 
클래스 작성하기

python 에서 자식 클래스 작성하기

class 자식클래스이름(부모클래스이름):

***********************************
  python 에서는 생성자도 상속 받음
***********************************

'''
# class Calc4_2 extends Calc4 (in JAVA)
class Calc4_2(Calc4):  
    def power(self):
        return self.num1 ** self.num2
```

```python
c1_2 = Calc4_2(10, 8)
print(c1_2.num1)
print(c1_2.num2)

print(c1_2.add())
print(c1_2.subtract())
print(c1_2.multiply())
print(c1_2.divide())
print(c1_2.power())
```

```python
class Animal:
    def sound(self):
        print('소리를 냅니다')
class Cat(Animal):
    # method overriding
    # 부모클래스의 메소드를 상속 받아서
    # 선언부는 동일하게 작성하고
    # 구현부(body)를 자식클래스에 맞도록 변형해서 작성함
    # (python 에서는 method overloading 은 없음)
    def sound(self):
        print('소리를 냅니다. 야옹~~~')
class Dog(Animal):
    def sound(self):
        print('소리를 냅니다. 멍멍~~~')
a1 = Animal()
a1.sound()
c1 = Cat()
c1.sound()
d1 = Dog()
d1.sound()
```

```python
'''
polymorphism (다형성)
  동일한 형태의 실행 code 가
  program 실행시 어떤 객체에 의해서
  호출되느냐에 따라서
  다른 결과가 실행되는 것
    ㄴ 상속관계에서 overriding 이 
       구현되어 있어야 구현할 수 있음
'''
# a1 - Animal, c1 - Cat, d1 - Dog
animals = [a1, c1, d1]
for animal in animals:
    animal.sound()
```

```python
class Parent:
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def display(self):
        print(f'이름 : {self.name}, 나이 : {self.age}')
p1 = Parent('부모', 62)        
p1.display()
```

```python
class Child(Parent):
    def __init__(self, name, age, height):
        self.name = name
        self.age  = age
        self.height = height
    def display(self):
        print(f'이름 : {self.name}, 나이 : {self.age}, 키 : {self.height}')       
ch1 = Child('자식', 24, 197)        
ch1.display()
```

```python
class Child2(Parent):
    def __init__(self, name, age, height):
        super().__init__(name, age)
        self.height = height
    def display(self):
        super().display()
        print(f'키 : {self.height}')       
ch2 = Child2('자식', 24, 197)        
ch2.display()
```

```python
class Child3(Parent):
    def __init__(self, name, age, height):
        Parent.__init__(self, name, age)
        self.height = height
    def display(self):
        Parent.display(self)
        print(f'키 : {self.height}')       
ch3 = Child3('자식', 24, 197)        
ch3.display()
```
