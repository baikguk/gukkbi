```python
'''
리스트 내포에서 if ~ else 사용하기
  ㄴ if ~ else 구문은 for 문 앞에 작성함
'''

'''
비어있는 리스트를 생성하고
0 부터 9 까지 의 숫자 중에서
짝수는 제곱하고 홀수는 세제곱해서 추가하세요
1) 일반 for 문 
2) 리스트 내포
'''
list1 = []
for number in range(10):
    if not number % 2:
        list1.append(number ** 2)
    else:
        list1.append(number ** 3)
print('list1 :',list1)        
print([number**2 if not number % 2 else number ** 3 for number in range(10)])
```

```python
'''
비어 있는 리스트를 생성하고
names 에서 이름은 대문자로 변환해서 추가하고
None 은 빈문자열로 추가하세요
1) 일반 for 문 
2) 리스트 내포
'''
names = ['tom', 'cruise', 'paul', 'jennifer', None, 'rara', None, 'kim']

list1 = []
for name in names:
    if name is not None:
        list1.append(name.upper())
    else:
        list1.append('')
print('list1 :',list1)
print([name.upper() if name is not None else '' for name in names])
```

```python
'''
비어 있는 리스트를 생성하고
nums 에서 45 이상인 수는 1을 더하고
45 미만인 수는 5를 더해서 추가하세요
1) 일반 for 문 
2) 리스트 내포
'''
nums = [22, 13, 45, 50, 98, 68, 42, 47, 3]

list1 = []
for number in nums:
    if number >= 45:
        list1.append(number + 1)
    else:
        list1.append(number + 5)
print('list1 :',list1)     

# 리스트 내포
print([number + 1 if number >= 45 else number + 5 for number in nums])
```

```python
'''
비어 있는 리스트를 생성하고
name_age 에서 이름만 추출해서 <-- key 만 추출하라는 의미 
                                  keys() 를 호출해도 되고 
                                  dictionary 이름만 사용해도 됨
                                  value 를 추출할 때는 
                                  반드시 values() 를 호출해야 함
비어 있는 리스트에 추가하세요
'''
name_age = {'이순신':46, '강감찬':72, '최영':64, '안중근':30, '이준':47}
list1 = []
for name in name_age.keys():
    list1.append(name)
print('list1 :',list1)  
# dictionary 이름만 사용해도 되고
print([name for name in name_age])
#  keys() 를 호출해도 됨
print([name for name in name_age.keys()])

'''                                         ┌ value 를 추출하라는 의미
비어 있는 리스트를 생성하고 name_age 에서 나이만 추출해서 추가하세요
'''
ages = []
for age in name_age.values():
    ages.append(age)
print('ages :',ages)    
print([age for age in name_age.values()])
```

```python
'''
items() 메소드를 사용해서 이름, 나이 추가하기
for name, age 에서 name, age 은 tuple 의 ( ) 가 생략된 형태임
  ㄴ list1.append((name, age)) 에서 (name, age) 처럼 괄호로 묶어줘야 함
'''
list1 = []
for name, age in name_age.items():
    list1.append((name, age))
print('list1 :',list1)
print([(name, age) for name, age in name_age.items()])
```

```python
'''
비어 있는 리스트를 생성하고
1 부터 20 까지의 숫자 중에서
3 의 배수는 제외하고 추가하세요
'''
list1 = []
for number in range(1, 21):
    # if number % 3 != 0
    if number % 3:
        list1.append(number)
print('list1 :',list1)
print([number for number in range(1, 21) if number % 3])
```

```python
'''
비어 있는 리스트를 생성하고 
1 부터 20 까지의 정수 중에서 <-- for number in range(1, 21) 
3 의 배수를 제외하고    <-- 전제 조건 - 외부 if 문
모든 숫자들을 추가하되  <-- 하되 : and 조건 - 중첩 if 문으로 처리함
2 의 배수는 음수로 만들어서 추가하세요  <-- 내부 if 문
 ㄴ 2 의 배수가 아닌 숫자들은 그대로 추가하라는 의미
'''
list1 = []
for number in range(1, 21):
    if number % 3:
        if not number % 2:
            list1.append(-number)
        else:
            list1.append(number)
print('list :',list1)   

'''
리스트 내포로 작성하기
1) 외부 if 문은 for 문 뒤에 작성하고
2) 내부 if 문은 for 문 앞에 작성함
'''
print([-number if not number % 2 else number for number in range(1, 21) if number % 3])
```
