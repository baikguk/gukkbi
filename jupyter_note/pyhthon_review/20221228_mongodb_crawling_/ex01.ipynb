```python
import json
import pprint

data = '''{
    "id":"01", "language":"Java", "edition":"third", "author":"Herbert Schildt",
    "history":
    [{
        "date":"2015-03-11",
        "item":"iPhone"
     },
     {
         "date":"2016-03-11",
         "item":"android"
     }
    ]
}'''

'''
json.loads(data) <-- json 모듈의 loads() 함수는
                     json 형태 (문자열) 의 data 를 
                     dictionaty type 으로 변환시켜줌
                     
network 상에서 data 를 전송할 때는 json 형태로 
주고 받아야 data 대비 용량이 작아서 편리함
data 를 전송 받은 후에는 가공하기에 쉬운 형태인
python 의 dictionary 형으로 변환해서 사용함
'''
# data : <class 'str'>  <-- JSON 은 type 이 str type 임
print('data :',type(data))
dict_data = json.loads(data)
print('dict_data :',type(dict_data))
print('-- dict_data --')
pprint.pprint(dict_data)
print("dict_data['id'] :",dict_data['id'])
print("dict_data['language'] :",dict_data['language'])
print("dict_data['edition'] :",dict_data['edition'])
print("dict_data['author'] :",dict_data['author'])
print("dict_data['history'][0] :",dict_data['history'][0])
print("dict_data['history'][1] :",dict_data['history'][1])
print("dict_data['history'][0]['date'] :",dict_data['history'][0]['date'])
print("dict_data['history'][1]['date'] :",dict_data['history'][1]['date'])
print("dict_data['history'][0]['item'] :",dict_data['history'][0]['item'])
print("dict_data['history'][1]['item'] :",dict_data['history'][1]['item'])
```

```python
import requests
import openpyxl
import re

client_id = '네이버개발자센터아이디'
client_secret = '네이버개발자센터비번'

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

# 두 번째와 세 번째 항목은
# title 과 link 라서 크기를 크게 설정함
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
# 맨 위 첫 줄에 column 제목 설정하기
excel_sheet.append(['랭킹', '제목', '링크'])


number, start = 0, 1
for index in range(10):
    start_num = start + (index * 100)
    # goods = input('검색할 상품을 입력하세요 : ')
    naver_open_api = f'https://openapi.naver.com/v1/search/shop.json?query=android&display=100&start=1'
    header_params = \
      {'X-Naver-Client-id':client_id, 'X-Naver-Client-Secret':client_secret}

    response = requests.get(naver_open_api, headers=header_params)
    # print('response.status_code :',response.status_code)

    if response.status_code == 200:
        data = response.json()
        for item in data['items']:
            number += 1
            # jupyter notebook 콘솔창에 출력
            # print(number,'번째 :',item['title'].replace('<b>','').replace('</b>',''),item['link'])        
            # excel 파일에 추가
            item_title = re.sub('\<\w*\>','',item['title'])
            item_title = re.sub('\</\w*\>','',item_title)
            # print(idx+1,'번째 :',item_title,item['link'])  
            excel_sheet.append([number,item_title,item['link']])
    else:
        print('Error Code :',response.status_code)

# 엑셀 파일 저장하기       
excel_file.save('shopping_list.xlsx')
excel_file.close()
```
