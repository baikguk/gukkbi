```python

# 대기오염정보


from bs4 import BeautifulSoup # 분석하기 용이하게 바꿔줌
import requests
import warnings
warnings.filterwarnings('ignore')
import openpyxl

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

# excel_sheet.column_dimension['B'].width = 
excel_sheet.append(['index','날짜','오염등급'])

url= f'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getMsrstnAcctoRltmMesureDnsty?stationName=종로구&dataTerm=month&pageNo=1&numOfRows=100&returnType=xml&serviceKey=LX6dNOD2rgJ1GpzPBf4KuwSUul9dfILpNZoOiiaWx8e0o077VN5mgs8xElRtC0Dx1XETQkILzF1nNm6HaTWkJw%3D%3D'
response = requests.get(url)
        #┌ resoponse.content만 봐도 분석하기 어지러우니 뷰티펄수프를 통해 형태로 바꿔줌
soup = BeautifulSoup(response.content,'html.parser')
data = soup.find_all('item')
# print(len(data))

for idx,i in enumerate(data):
    datatime=i.find('datatime')
    pm10grade = i.find('pm10grade')
    excel_sheet.append([idx,datatime.get_text(),pm10grade.get_text()])
excel_file.save('날짜 및 시간별 대기오염등급.xlsx')
excel_file.close()
```
